```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Function representing the heat conduction equation
def heat_conduction(y, x, k):
    return -k * y

# Solving the heat conduction equation using odeint
def solve_heat_transfer(k, initial_temp, length, num_points):
    x = np.linspace(0, length, num_points)
    y0 = initial_temp
    solution = odeint(heat_conduction, y0, x, args=(k,))
    return x, solution

# Plotting the temperature distribution
def plot_temperature_distribution(x, temperature_profile):
    plt.plot(x, temperature_profile, label='Temperature Profile')
    plt.xlabel('Distance')
    plt.ylabel('Temperature')
    plt.title('Heat Transfer - 1D Conduction')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    # Parameters
    thermal_conductivity = 0.01  # Example value, replace with the actual value for your case
    initial_temperature = 100.0  # Example value, replace with the actual value for your case
    length_of_rod = 10.0  # Example value, replace with the actual value for your case
    num_points = 100

    # Solve heat transfer problem
    x_values, temperature_profile = solve_heat_transfer(thermal_conductivity, initial_temperature, length_of_rod, num_points)

    # Plot the results
    plot_temperature_distribution(x_values, temperature_profile)
```
